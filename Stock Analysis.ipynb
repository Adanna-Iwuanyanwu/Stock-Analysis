```python
#import the libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#the start and end date for the required stocks
start_date = '2013-01-01'
end_date = '2023-03-31'
```

```python
#stock data
tech_coms = ['MSFT', 'GOOGL', 'AAPL', 'META']
```

```python
#loading the stock data using globals() as a simple way of setting the DataFrame.
for stock in tech_coms:
    globals()[stock] = yf.download(stock, start_date, end_date)
    
com_list = [AAPL, GOOGL, MSFT, META]
Company_Names = ["APPLE", "GOOGLE", "MICROSOFT", "META"]

for stock_data, names in zip(com_list, Company_Names):
    stock_data['Company_Names'] = names
    
stock_data = pd.concat(com_list, axis = 0)
```

```python
#checking the data
stock_data.head(2)
```

```python
#data info
stock_data.info()
```

```python
#Stock price trend over time
plt.figure(figsize = (15, 8))
plt.title('Stock Price Over Time', color = 'black', fontsize = 20)

AAPL['Open'].plot(label = 'APPLE')
MSFT['Open'].plot(label = 'MICROSOFT')
META['Open'].plot(label = 'META')
GOOGL['Open'].plot(label = 'GOOGLE')

plt.legend()
plt.show()
```

The graph shows the price trend of the stocks as the year goes by. In 2021 and 2022, the stock market was booming as the various companies experienced a great peak in stock price. This first quarter of 2023, the stock market has been in bad shape as the stock price keeps showing a downtrend in its trade. 

Comparing the four stocks with the current market trend, Microsoft stock is doing better in the stock market than the other stocks with Google having a serious downtrend in sales. Although, there has been a drastic downtrend of Meta stocks in the market currently.

```python
#Stock purchase over time
plt.figure(figsize = (15, 8))
plt.title('Stock Purchase Over Time', color = 'black', fontsize = 20)

AAPL['Volume'].plot(label = 'APPLE')
MSFT['Volume'].plot(label = 'MICROSOFT')
META['Volume'].plot(label = 'META')
GOOGL['Volume'].plot(label = 'GOOGLE')


plt.legend()
plt.show()
```

Over the years, Apple stock has been purchased more by investors. But, there has been a drastic reduction in stock purchase this year (2023) compared to other years. Even with the downtime in stock purchase, Apple stock seems to have traded more in the market than other stocks.

```python
#market cap
plt.figure(figsize = (15, 8))
plt.title('Market Capitalisation', color = 'black', fontsize = 20)

AAPL['market_cap'] = AAPL['Open'] * AAPL['Volume']
MSFT['market_cap'] = MSFT['Open'] * MSFT['Volume']
META['market_cap'] = META['Open'] * META['Volume']
GOOGL['market_cap'] = GOOGL['Open'] * GOOGL['Volume']

AAPL['market_cap'].plot(label = 'APPLE')
MSFT['market_cap'].plot(label = 'MICROSOFT')
META['market_cap'].plot(label = 'META')
GOOGL['market_cap'].plot(label = 'GOOGLE')

plt.legend()
plt.show()
```

The graph shows Apple company has more worth than the rest of the other companies. Even with the fluctuation in stock price in the first quarter of 2023, the market capitalisation of the companies have seems to be on the rise compared to the previous years.

```python
#movig average
MA50 = AAPL['Close'].rolling(50).mean()
MA50_1 = GOOGL['Close'].rolling(50).mean()
MA50_2 = META['Close'].rolling(50).mean()
MA50_3= MSFT['Close'].rolling(50).mean()

MA200 = AAPL['Close'].rolling(200).mean()
MA200_1 = GOOGL['Close'].rolling(200).mean()
MA200_2 = META['Close'].rolling(200).mean()
MA200_3 = MSFT['Close'].rolling(200).mean()
```

```python
MA_50 = pd.DataFrame({'APPLE': MA50, 'GOOGLE':MA50_1, 'META':MA50_2, 'MICROSOFT':MA50_3})
MA_200 = pd.DataFrame({'APPLE': MA200, 'GOOGLE':MA200_1, 'META':MA200_2, 'MICROSOFT':MA200_3})

MA_50.plot(figsize=(15, 8), title = 'MOVING AVERAGE 50')
MA_200.plot(figsize=(15, 8), title = 'MOVING AVERAGE 200')

plt.legend()
plt.show()
```

The 50-days and 200-days moving average shows that Microsoft traded well in the stock market in the first quarter of 2023 beacuase it was above the other stock closing prices while Google traded the least in the stock market.
